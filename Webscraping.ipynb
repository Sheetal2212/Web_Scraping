# Extraction of latitude-longitude information associated with different markers 

# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

# Loading File

```python
js = open("C:\\Users\\HP\\Downloads\\BMC-ContainmentZones.js", "r").readlines()
json_str = json. dumps(js)
```

```python
soup = BeautifulSoup(json_str, 'html.parser')
#soup
```

```python
tags = soup.find_all('script')
t= tags[8] #obtaining script tag that contains the information to be exctracted
```

```python
type(t)
```

```python
scripts = []
for x in t:
    scripts.append(str(x)) #converting bs4.element.Tag to list for further processeing
```

```python
l= ' '.join([str(elem) for elem in scripts]) #converting list to string using list comprehension
l=l.split('\\n') # spliting to form list
l1 = ' '.join([str(elem) for elem in l])
l1= l1.split('", "') 
l1 = [i for i in l1 if i.strip()] #removing whitespaces
l1
```

```python
l1= ' '.join([str(elem) for elem in l1]) 
l1=l1.split(';')
l1
```

```python
df = pd.DataFrame(l1) #converting l1 to data frame
df.head()
```

```python
pd.set_option('display.max_colwidth', -1) # to display complete content of all columns
```

```python
df= df[df[0].str.contains("var marker_")] # extracting only those columns that starts with var marker_
df
```

```python
df= df[0].str.split(']',expand=True)
df
```

```python
df.drop(1,axis=1,inplace=True)
df
```

```python
df= df[0].str.split('[',expand=True)
df
```

```python
lmarks= df[1]
lmarks= pd.DataFrame(lmarks)
lmarks= lmarks[1].str.split(',',expand=True)
lmarks=lmarks.rename(columns={0:'latitude',1:'longitude'})
```

```python
lmarks
```

```python
mark_id = df[0].str.split('_',expand=True)
mark_id
```

```python
mark_id = mark_id[1].str.split('=',expand=True)
mark_id
```

```python
mark = mark_id.drop(1,axis=1)
mark
```

```python
mark=mark.rename(columns={0:'marker_id'}) 
mark
```

```python
final = [mark,lmarks]
result = pd.concat(final, axis=1)
result # final data frame containg latitude and longitude associated with different markers
```
